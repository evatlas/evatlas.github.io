```python
import requests
import json
from pathlib import Path

from dotenv import load_dotenv
import os
```

```python
RAW_DIR = Path.cwd().parent / "data" / "raw"
print(RAW_DIR)
```

```python
# Load the API KEY from the .env file
load_dotenv() # this reads the .env file
key = os.getenv("NRL_API_KEY")
if key:
    print(key[:5], "---") # print first 5 characters
else:
    print("NRL_API_KEY not found in .env file")
```

```python
# documentation
# https://developer.nlr.gov/docs/transportation/alt-fuel-stations-v1/all/#request-url
```

```python
# set parameters for the API request
url = "https://developer.nlr.gov/api/alt-fuel-stations/v1.json" # Website URL for the API endpoint

api_key = key # API KEY from .env file
fuel_type = "ELEC"  # Electric vehicle charging stations
country = "CA"  # Canada
limit = "all"  # Retrieve all stations

# combine for the request
params = {
    "api_key": api_key,
    "fuel_type": fuel_type,
    "country": country,
    "limit": limit
}
```

```python
# Retreive the data from the API
response = requests.get(url, params = params)
print(response.status_code)  # Print the status code to check if the request was successful --- 200 means success
```

```python
# Quick data check
data = response.json()  # Convert the response to JSON format
print(data.keys())  # Print the keys of the JSON data to understand its structure
print(len(data['fuel_stations']))  # Print the number of fuel stations retrieved
print(data['fuel_stations'][0])  # Print the first fuel station's data to see its structure
```

```python
# better format data view check
print(json.dumps(data["fuel_stations"][0], indent=2))
```

```python
# save data
out_path = RAW_DIR / "chargers.json"
out_path.parent.mkdir(parents=True, exist_ok=True)
with open(out_path, "w", encoding = "utf-8") as f:
    json.dump(data, f)
```

```python
# check the saved file
with open(out_path) as f:
    check = json.load(f)
print(check.keys())  # Print the keys of the loaded JSON data to verify it was saved correctly
print(len(check["fuel_stations"]))  # Print the number of fuel stations in the saved file
```
