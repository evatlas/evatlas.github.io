```python
import requests
import zipfile
from pathlib import Path
```

```python
# Where raw files go
RAW_DIR = Path.cwd().parent / "data" / "raw"
print(RAW_DIR)
```

```python
# Table 20-10-0025-01
PRODUCT_ID = "20100025"
```

```python
# Retreive data
```

```python
# Step 1: ask WDS for the download link
url = f'https://www150.statcan.gc.ca/t1/wds/rest/getFullTableDownloadCSV/{PRODUCT_ID}/en'
response = requests.get(url)
response.raise_for_status()  # Check for HTTP request errors
print(response.json())

zip_url = response.json()['object']
print(zip_url)
```

```python
# Step 2: download the zip file
data = requests.get(zip_url)
data.raise_for_status()  # Check for HTTP request errors
```

```python
# Step 3: save the zip file to disk
zip_path = RAW_DIR / f"{PRODUCT_ID}.zip"
zip_path.parent.mkdir(parents=True, exist_ok=True)
zip_path.write_bytes(data.content)
```
